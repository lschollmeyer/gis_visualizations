```python
import folium
import pandas as pd
from IPython.display import HTML

def display(m, height=300):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{0}" '
                 'style="width: 100%; height: {1}px; '
                 'border: none"></iframe>'.format(srcdoc, height))
    return embed

```

```python
ufo = pd.read_csv('data/ufo_geometry.csv')
ufo.dropna(inplace=True)
ufo.head()
```

```python
map = folium.Map(location=BLADEN, zoom_start=6)

marker_cluster = folium.MarkerCluster("ufo_cluster").add_to(map)

for i,row in ufo[ufo['State'] == 'ne'].iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Shape'] + ' - ' + row['Summary']).add_to(marker_cluster)

map
```

```python
ufo_counts = pd.DataFrame(ufo.groupby('State').size().reset_index())
ufo_counts.columns = ['abbr', 'ufo_count']
ufo_counts['abbr'] = ufo_counts['abbr'].str.upper()
ufo_counts.head()
```

```python
map1 = folium.Map(location=[48, -102], zoom_start=3)
map1.choropleth(geo_path= 'data/us_states.json', data_out='data.json', data=ufo_counts,
             columns=['abbr', 'ufo_count'],
             key_on='feature.id',
             fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2)

map1
```

```python
LINCOLN = (40.8258, -96.6852)
```

```python
import json

with open('data/bike-paths.geojson') as f:
    data = json.load(f)

path_data = []
for feature in data['features']:
    path_data.append(feature['geometry']['coordinates'])
```

```python
test = []

for i in path_data:
    for j in i:
        inner_test = []
        for k in j:
            inner_test.append([k[1], k[0]])
        test.append(inner_test)
```

```python
map_paths = folium.Map(location=LINCOLN, zoom_start=10)
for x in test:
    folium.PolyLine(x, weight=8).add_to(map_paths)

map_paths
```
